# Data Cleaning & Casting: Professional Standards

### 1. Handling Missing Values (`NaN`)
* **`np.nan` vs `None`**: In Pandas, `np.nan` is a float-type value. This is why a column containing missing numeric data is often cast to `float`.
* **Standardizing Missing Data**: Use `na_values` during `read_csv()` to catch custom missing strings (like 'NA', 'Missing', 'Anonymous') early in the pipeline.

### 2. Dropping & Filling (`dropna` / `fillna`)
* **Strategic Dropping**:
    * `how='all'`: Removes records only if every column is empty.
    * `subset=['col']`: Target critical columns. If essential data is missing, the row is discarded.
* **Imputation (`fillna`)**: Filling missing values with a default (like `0` or `"Missing"`) helps keep the dataset's shape while ensuring logic doesn't break during calculations.

### 3. Type Casting (`astype`)
* **The `float` Necessity**: You cannot calculate the mean of a column that contains `NaN` unless its type is `float`.
* **Data Transformation**: Always check `df.dtypes` before performing mathematical operations.
* **Custom Replacement**: Use `.replace()` to convert categorical strings (e.g., 'Less than 1 year') into numeric equivalents before casting.

### 4. ML Context: Data Quality
* **Garbage In, Garbage Out**: Clean data is more important than the ML algorithm itself. Handling outliers and missing values directly prevents "Bias" and errors in model training.

```python
import pandas as pd
import numpy as np

people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com',
              'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)

df.replace("NA", np.nan, inplace=True)
df.replace("Missing", np.nan, inplace=True)
```

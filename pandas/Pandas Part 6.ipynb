Structural Data Engineering (Add/Remove)
1. Feature Engineering Foundations
Constructing New Features: Combining columns (e.g., First + Last) to create Full_Name is the first step in Feature Engineering, transforming raw data into meaningful inputs for ML models.

Vectorized Operations: Always prefer direct operations like df['A'] + df['B']. They are significantly faster than Python loops because they leverage NumPy's vectorization under the hood.

2. Advanced String Splitting
The expand=True Argument: This is critical for returning a new DataFrame structure instead of a Series of lists, ensuring your table remains consistent for further analysis.

The n Parameter: This defines the number of splits.

Example: n=1 splits only at the first occurrence of a delimiter (like a space), which is essential for handling names with multiple middle parts while keeping only two resulting columns.

3. Efficient Data Cleaning & Maintenance
Filtering vs. Dropping:

While df.drop(index=...) is useful for specific row removal by label, it is often faster in large datasets to use Boolean Masking (the reverse filter) to redefine the DataFrame: df = df[df['Gender'] != 'Male'].

Memory & Inplace:

Using inplace=True or re-assigning the variable df = df.drop(...) ensures changes are committed to the memory reference of your object.

Column Reordering: You can reorder columns by passing a list of names in the desired order: df = df[['Col3', 'Col1', 'Col2']].

4. Row Concatenation (pd.concat)
Standard Method: pd.concat is now the standard for merging rows/DataFrames (replacing the deprecated .append()).

ignore_index=True: Always use this when the original row labels are not meaningful to prevent duplicated or confusing indices in your new dataset.

```python
# __ Add Remove Rows And Columns From DataFrame __
people = {
    "First name": ["John", "mark", "alexa"],
    "Last name": ["Doe", "Los", "kol"],
    "Age": [25, 30, 42],
    "Gender": ["Male", "Male", "Female"],
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df['First name'] + " " + df['Last name']  # add new column temporary
```

```python
df['Full_Name'] = df['First name'] + " " + df['Last name']  # add new column permanently
```

```python
df.drop(columns=["First name", "Last name"])  # remove new column temporary
```

```python
df = df.drop(columns=["First name", "Last name"])  # remove new column permanently
```

```python
# df.drop(columns=["First name","Last name"],inplace=True)
# same to df = df.drop(columns=["First name","Last name"])
```

```python
# df["Full_Name"].str.split() return a series of list ....
df["Full_Name"].str.split(" ", expand=True)  # return a split dataframe split (temporary)
```

```python
df[["First", "Last"]] = df["Full_Name"].str.split(" ", expand=True)  # permanently
```

```python
new_row = pd.DataFrame({"Age": [50, 60, 70, 30, 20]})
df = pd.concat([df, new_row], ignore_index=True)
# concat tow dataframe
```

```python
new_row = pd.DataFrame({"Gender": ["Male"] * 6})
df = pd.concat([df, new_row], ignore_index=True)
```

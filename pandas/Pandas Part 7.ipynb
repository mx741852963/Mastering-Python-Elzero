# Data Sorting & Ranking Insights

### 1. Multi-Index Sorting
* **Primary & Secondary**: When sorting by multiple columns `by=['Col1', 'Col2']`, the second column acts as a tie-breaker for the first.
* **Directional Control**: Use `ascending=[True, False]` to sort categories alphabetically while keeping numerical values in descending order simultaneously.

### 2. High-Performance Selection (`nlargest` / `nsmallest`)
* **Efficiency**: Use `.nlargest(n, 'column')` instead of `.sort_values().head(n)`.
* **Mechanism**: It uses a heap-based selection which is computationally cheaper than a full sort ($O(n \log k)$ vs $O(n \log n)$).
* **Usage**: Works on both DataFrames (returns full rows) and Series (returns only values).

### 3. Index Integrity
* **Sorting Side-Effects**: Sorting shuffles the row labels (Index).
* **Resetting**: Use `df.reset_index(drop=True)` to re-establish a standard 0-based sequence after a permanent sort.

### 4. ML Context: Outlier Detection
* Sorting is a key part of **Exploratory Data Analysis (EDA)**. By checking the `nlargest` and `nsmallest` values, we can quickly identify **Outliers** (extreme data points) that might negatively impact Machine Learning model accuracy.
